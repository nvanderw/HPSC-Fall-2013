#Simpson's Rule

```python
import math

def func_a(x):
    x = x-10
    return 0.25*x*x*x + 0.75*x*x -1.5*x + 8

def func_b(x):
    return 10000.0*math.exp(-x)/float(x)

def func_c(x):
    return sin(x)*x

def func_d(x):
    x = x-7.5
    return x/sqrt(1 + x*x) + 2.0
```

```python
import matplotlib.pylab as plt
import numpy as np

x = np.linspace(5,10,100)
y_a = map(func_a, x)
y_b = map(func_b, x)
y_c = map(func_c, x)
y_d = map(func_d, x)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(x, y_a)
ax[0,1].plot(x, y_b)
ax[1,0].plot(x, y_c)
ax[1,1].plot(x, y_d)
```

##What is the area under the curve?

$$
\int_a^b f(x) dx \approx \frac{b-a}{6} \left[ f(a) +4f\left[\frac{a+b}{2}\right] + f(b) \right]
$$


```python
def simpsons(x, f):
    a = min(x)
    b = max(x)
    return (b-a)/6.0 * (f(a) + 4*f((a+b)*0.5) + f(b)) 
```

```python
area_a = simpsons(x,func_a)
area_b = simpsons(x,func_b)
area_c = simpsons(x,func_c)
area_d = simpsons(x,func_d)

print area_a
print area_b
print area_c
print area_d
```

## Good or Bad? 

- Break the problem into $n$ parts.
- Divide the area under the curve into n equal parts.

$$
\int_a^b f(x) dx \approx \frac{\Delta x}{3}(y_o +4y_1 +2y_2+4y_3+2y_4+ \cdots +2y_{n-2}+4y_{n-1}+y_n)
$$

NOTE: $n$ must be an even number.

```python
# From Wikipeia
def simpson_iter(f, a, b, n):
    """Approximates the definite integral of f from a to b by
    the composite Simpson's rule, using n subintervals"""
    h = (b - a) / float(n)
    s = f(a) + f(b)
    for i in range(1, n, 2):
        s += 4 * f(a + i * h)
    for i in range(2, n-1, 2):
        s += 2 * f(a + i * h)
 
    return s * h / 3
```

```python
print simpson_iter(func_a, 5, 10, 2)
print simpson_iter(func_a, 5, 10, 10)
print simpson_iter(func_a, 5, 10, 100)
print simpson_iter(func_a, 5, 10, 1000)
```

```python
print simpson_iter(func_b, 5, 10, 2)
print simpson_iter(func_b, 5, 10, 10)
print simpson_iter(func_b, 5, 10, 100)
print simpson_iter(func_b, 5, 10, 1000)
print simpson_iter(func_b, 5, 10, 10000)
```
