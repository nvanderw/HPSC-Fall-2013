## Use `pandas`

```python
import pandas as pd
grades = pd.read_csv('grades_ordered.csv')
print grades.head()
```

## Plot Total

```python
fig, ax = plt.subplots(1)

ax.set_title('Grades for assigment one')

ax.plot(grades.index, grades['Total'], linewidth=1.0)
ax.scatter(grades.index, grades['Total'])

ax.set_xlabel('anonymous student')
ax.set_ylabel('total points')
plt.show()
```

## Multiple lines

```python
fig, ax = plt.subplots(1)

ax.set_title('Grades for assigment one')

ax.plot(grades.index, grades['Total'], linewidth=1.0, label="Total")
ax.plot(grades.index, grades['Code'], linewidth=1.0, label="Code")
ax.plot(grades.index, grades['Makefile'], linewidth=1.0, label="Makefile")
ax.plot(grades.index, grades['Plot'], linewidth=1.0, label="Plot")

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc=2)

ax.set_xlabel('anonymous student')
ax.set_ylabel('total points')

plt.ylim([-5,55])
plt.show()
```
